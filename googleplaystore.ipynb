```python
#Thinking Process:
item="19M"
item[-1]
```

```python
#1 MB=1024 Kb

def size_process(item):
    if str(item)[-1] == "M":
        res = float(str(item).replace("M", ""))
        res = res * 1024
        return res
    elif str(item)[-1] == "k":
        res = float(str(item).replace("k", ""))
        return res
    else:
        return str(np.nan)
```

```python
size_process("19M")
```

```python
df_copy["Size"] = df_copy["Size"].apply(size_process)
```

```python
#Successfully converted all values to a single unit.
df_copy["Size"].unique()
```

#Univariate Analysis

```python
#categorical feature analysis
df_copy["Type"].value_counts(normalize=True)*100
```

*Insights: 92% Apps are free, making it accessible to wide number of people.*

```python
plt.figure(figsize=(8,5))
ax = sns.countplot(x='Content Rating', data=df, palette='plasma')

# total count
total = len(df)

# add percentage labels
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom')

plt.title("Distribution of Content Rating")
plt.ylabel("Count")
plt.show()
```

*Insights: Adults only 18+ and unrated movies are being given the least ratings.*

```python
# Get percentage distribution
android_pct = df_copy["Android Ver"].value_counts(normalize=True) * 100

# Keep top 5 and group the rest as "Others"
top_n = 10
android_top = android_pct[:top_n]
android_top["Others"] = android_pct[top_n:].sum()

# Plot pie chart
android_top.plot(
    kind="pie",
    autopct="%.1f%%",
    figsize=(6,6),
    ylabel="",
    title="Top 10 Android Versions Distribution"
)
```

*Insights: Most apps are updated to 4.0.3, while only 5.3% apps are updated to the latest 5.0 version.*
